```python
#使うライブラリ
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
df_hist = pd.read_csv('matplotlib_quiz/hist_data.csv')
plt.hist(df_hist['aデータ'], alpha = 0.5, bins=30, label="aデータ")
plt.hist(df_hist['bデータ'], alpha = 0.5, bins = 30, label="bデータ")
plt.legend(loc='upper right', fontsize=10)

plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
df_scatter =  pd.read_csv('matplotlib_quiz/scatter_data.csv')
df_class0 = df_scatter[(df_scatter['class']==0)]
df_class1 = df_scatter[(df_scatter['class']==1)]
df_class2 = df_scatter[(df_scatter['class']==2)]
plt.scatter(df_class0['length (cm)'], df_class0['width (cm)'], label="class0") 
plt.scatter(df_class1['length (cm)'], df_class1['width (cm)'], label="class1") 
plt.scatter(df_class2['length (cm)'], df_class2['width (cm)'], label="class2") 
plt.title("クラス０〜２のlengthやwidthに関する散布図")
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.legend(loc='upper right', fontsize=10)
plt.grid()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
df_line = pd.read_csv('matplotlib_quiz/line_data.csv')
plt.plot(df_line['期間'], df_line['件数'], marker='o')
plt.grid()
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")

```

```python
plt.plot(df_line['期間'], df_line['件数'])

plt.bar(df_line['期間'], df_line['費用'])
plt.grid()
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")

```
